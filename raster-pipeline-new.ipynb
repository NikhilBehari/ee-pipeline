### Import required packages

```python
from osgeo import gdal, osr, ogr
import matplotlib.pyplot as plt
from osgeo.gdalconst import *
import numpy as np
import os
import osmnx as ox
import rasterio
import geopandas as gpd
from shapely.geometry import Polygon
import json
import math
```

```python
# CONSTANTS 
input_dir = "input/"
inter_dir = "temp/"
output_dir = "output/"
target_ext = []
target_res = []
```

```python
def getExtent(path):
    raster_file = gdal.Open(path,0)
    geoTransform = raster_file.GetGeoTransform()
    minx = geoTransform[0]
    maxy = geoTransform[3]
    maxx = minx + geoTransform[1]*raster_file.RasterXSize
    miny = maxy + geoTransform[5]*raster_file.RasterYSize
#     return [minx,miny,maxx,maxy]
    return [minx,maxy,maxx,miny]
```

```python
def getResolution(path):
    reference = gdal.Open(path, 0)
    rt = reference.GetGeoTransform()
    xr = rt[1]
    yr = -rt[5]
    return [xr, yr]
```

```python
# File to copy extent and resolution 
ref_file = "input/sentinel.tif"
target_ext = getExtent(ref_file)
target_res = getResolution(ref_file)
print("Target Extent: " + str(target_ext))
print("Target Resolution: " + str(target_res))
```

```python
if not os.path.exists(inter_dir):
    os.mkdir(inter_dir)
    
# Iterate through input directory and find .tif files
input_tifs = []
for subdir, dirs, files in os.walk(input_dir):
    for f in files:
        path = subdir + os.sep + f
        if path.endswith(".tif"):
            input_tifs.append(path)
            
# Resize, resample and save to intermediate dir
for tif in input_tifs:
    inputDs = gdal.Open(tif)
    
    ds = gdal.Warp(str(inter_dir + os.path.basename(tif)), inputDs,
                   xRes=target_res[0], yRes=target_res[1],
                   outputBounds=target_ext,
                   resampleAlg= gdal.GRIORA_Bilinear,
    #                format='MEM'
                   format='GTiff'
                  )
    ds = None
```

```python
# Iterate through intermediate product directory
tifs = []
numLayers = 0
for subdir, dirs, files in os.walk(inter_dir):
    for f in files:
        path = subdir + os.sep + f
        if path.endswith(".tif"):
            tifs.append(path)
            temp_rast = rasterio.open(path)
#             print(gdal.Info(inputDs))
            numLayers += temp_rast.count
  
reference = rasterio.open(tifs[0])
meta = reference.meta
meta.update(count = numLayers, dtype=rasterio.float32)

with rasterio.open("output/stacked.tif", "w", **meta) as dest: 
    for tif in tifs:
        temp_raster = rasterio.open(tif) 
        for ind in range(temp_raster.count):
            ind += 1 # Raster layers start at 1
            raster_layer = temp_raster.read(ind)
            raster_layer = raster_layer.astype('float32')
            dest.write_band(ind, raster_layer)
```
